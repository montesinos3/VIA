# Instalación del entorno miniforge en Mac de la serie M
En el caso de la nueva versión M1 de Mac, vamos a usar una instalación de Miniforge, que es una alternativa a miniconda con énfasis en la compatibilidad con la nueva arquitectura de apple.

Lo descargamos del siguiente enlace:

https://github.com/conda-forge/miniforge

En nuestro caso descargamos la versión *Miniforge3-MacOSX-arm64*, y una vez descargada la instalamos con

```
bash Miniforge3-MacOSX-arm64
``` 

y seguimos las instrucciones. Una vez instalado, creamos un entorno de conda

```
conda create --name via python==3.11
conda activate via
```

Y podemos instalar las librerías necesarias:

```
pip install -r requirements_0.txt
```


## Instalar PyQT6

Para poder instalar PyQT hay que hacerlo con este comando:

```
pip install pyqt6
```

Además, con esta versión, en los programas donde se utiliza *pyqtgraph.Qt* con *QtGui*, hay que cambiar:
```
QtGui.QApplication.instance().exec_()
```
por:
```
QtWidgets.QApplication.instance().exec()
```

## Instalar Dlib

Para poder instalar Dlib, primero debemos instalar Cmake:
```
pip install cmake
```
Entonces descargamos Dlib:
```
pip install dlib
```

# Instalación de Tensorflow

```
python -m pip install tensorflow-macos
```

Después descargamos el plugin metal para Tensorflow, que acelera el funcionamiento de Tensorflow en este dispositivo, más info en: https://blog.tensorflow.org/2021/06/pluggabledevice-device-plugins-for-TensorFlow.html

```
python -m pip install tensorflow-metal
```

# Nota

Si al import cv2 salta un error referido a numpy hacer lo siguiente:
```
pip uninstall numpy    # Ejecutar este comando dos veces seguidas
pip install numpy
```
